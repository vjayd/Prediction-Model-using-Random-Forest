```python
# pandas
import pandas as pd
from pandas import Series,DataFrame

# numpy, matplotlib, seaborn
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

# machine learning
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier



#Import training and test data into the data set
training_criminal = pd.read_csv("criminal/criminal_train.csv")
test_criminal = pd.read_csv("criminal/criminal_test.csv")






```

```python
#GET the overview of the data
training_criminal.head()
test_criminal.head()
```

```python
#Find out how many Nan values, object type of the data is there
training_criminal.info()
```

```python
#Find the correlation of the Criminal values with other values 
#and consider the values with positive correlation coefficient

corr_matrix = training_criminal.corr()
corr_matrix["Criminal"].sort_values(ascending=False)
```

```python
#Drop the columns having negative correlation coefficient as they will be an outlier to the given data

training_criminalx = training_criminal.drop(['PERID','IRFAMSVC','IRFAMSSI','IIWELMOS','VEREP','HLCALLFG',
                                            'HLCALL99','PRXYDATA','IRKI17_2','IFATHER','IRHHSIZ2','HLCNOTYR',
                                            'OTHINS','IRINSUR4','NRCH17_2','IRFAMSOC','IRPINC3','IRMEDICR',
                                            'POVERTY3','GRPHLTIN','IRFAMIN3','IRPRVHLT','Criminal'], axis=1)

training_criminaly = training_criminal['Criminal']

test_criminalx = test_criminal.drop(['PERID','IRFAMSVC','IRFAMSSI','IIWELMOS','VEREP','HLCALLFG',
                                            'HLCALL99','PRXYDATA','IRKI17_2','IFATHER','IRHHSIZ2','HLCNOTYR',
                                            'OTHINS','IRINSUR4','NRCH17_2','IRFAMSOC','IRPINC3','IRMEDICR',
                                            'POVERTY3','GRPHLTIN','IRFAMIN3','IRPRVHLT'], axis=1)

test_criminal.info()
```

```python
# Random Forests

random_forest = RandomForestClassifier(n_estimators=200)

random_forest.fit(training_criminalx,training_criminaly)

Y_pred = random_forest.predict(test_criminalx)

random_forest.score(training_criminalx,training_criminaly)
```

```python
submission = pd.DataFrame({
     "PERID": test_criminal["PERID"],
    "Criminal": Y_pred    
   
        
    })
submission.to_csv('csubmission.csv', index=False)
```
